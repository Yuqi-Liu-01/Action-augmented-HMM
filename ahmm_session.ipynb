```python
import pickle

from ahmm_utils import (
    SingleTrackConfig,
    SingleTrackGenerator,
    build_vocab,
    sweep_save_all_and_log,
)

```

```python
with open('sessions_combined.pkl', 'rb') as f:
    sessions_combined = pickle.load(f)

print(f"Loaded {len(sessions_combined)} sessions from 'sessions_combined.pkl'")
```

```python
heights_in_data = sorted({
    int(h) for sess in sessions_combined for (h, _, _) in sess["records"]
} | {
    int(h) for sess in sessions_combined for (_, h, _) in sess["records"]
})
heights = heights_in_data  # or overwrite with your canonical list if you prefer

vocab = build_vocab(heights, height_encoding='split')

# Your deterministic task config (you can change these to experiment)
cfg = SingleTrackConfig(
    tower_heights=heights,
    height_encoding='split',
    p_gap=1.0,
    p_reward_given_correct=1.0,
    p_reward_given_incorrect=0.0
)
gen = SingleTrackGenerator(vocab, cfg)

```

```python
states = 25
sweep_save_all_and_log(
    sessions_combined,
    vocab,
    gen,
    animals=("G375", "G386", "G402", "G410", "G420", "G506"),
    models_root=f"ahmm_models_per_session_all/{states}",
    seeds=range(0, 101),
    test_ratio=0.40,
    valid_ratio=0.7,
    n_states=states,
    max_iters=500000,
    train_kwargs=None,
    log_path=f"ahmm_models_per_session_all/{states}/model_log.pkl",
    checkpoint_every=1,
    overwrite=False,
    resume=True,
)
```
